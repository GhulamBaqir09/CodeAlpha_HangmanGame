```python
import random

# List of words for the game
word_list = ['python', 'hangman', 'programming', 'challenge', 'developer']

def hangman():
    # Select a random word from the word list
    word = random.choice(word_list)
    word_length = len(word)
    guessed_letters = set()
    incorrect_guesses = 0
    max_incorrect_guesses = 6  # Number of allowed incorrect guesses

    # Initialize display of the word with underscores
    display_word = ['_'] * word_length

    print("Welcome to Hangman!")
    print("Guess the word:")
    print(' '.join(display_word))

    while incorrect_guesses < max_incorrect_guesses and '_' in display_word:
        guess = input("Guess a letter: ").lower()

        # Check if the guessed letter is valid
        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input. Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You've already guessed that letter.")
            continue

        guessed_letters.add(guess)

        if guess in word:
            # Update the display word with the correct guess
            for index, letter in enumerate(word):
                if letter == guess:
                    display_word[index] = guess
            print(f"Good guess: {guess}")
        else:
            incorrect_guesses += 1
            print(f"Incorrect guess: {guess}. You have {max_incorrect_guesses - incorrect_guesses} guesses left.")

        print(' '.join(display_word))

    if '_' not in display_word:
        print(f"Congratulations! You've guessed the word: {word}")
    else:
        print(f"Game over! The word was: {word}")

if __name__ == "__main__":
    hangman()
```
